## 1. Sales



Fill in the TODO cells in sales.ipynb notebook.

```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('sales.csv')
df.head()
```

✓ Fix column datatypes.

```python
df["ordered_at"] = pd.to_datetime(df["ordered_at"], errors="coerce")
```

✓ Drop if duplicated or null.


```python
# Loại bỏ hàng trùng lặp
df = df.drop_duplicates()

# Loại bỏ hàng có giá trị null trong 'name' (hoặc có thể thay thế bằng giá trị khác)
df = df.dropna(subset=["name"])
```

✓ Sanity check for value ranges and to check assumptions.


```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
df['line_total'] = pd.to_numeric(df['line_total'], errors='coerce')


df = df[(df["price"] >= 0) & (df["quantity"] > 0) & (df["line_total"] >= 0)]
```

✓ Use regular expression and lambda function to parse data.


```python
df["name"] = df["name"].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True).str.strip().str.title()
# Check the result
df.info(), df.head()
```
